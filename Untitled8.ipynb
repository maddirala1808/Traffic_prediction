```python
# ==========================================================
# TRAFFIC STRESS PREDICTION USING LIGHTGBM
# PART 1
# ==========================================================

# Install required libraries
!pip install lightgbm category_encoders optuna -q

# ==========================================================
# Import Libraries
# ==========================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder

from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

from lightgbm import LGBMRegressor

import pickle
import optuna # Import Optuna

# Better display

pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',100)

print("Libraries Loaded Successfully")
from google.colab import files
import zipfile

uploaded = files.upload()

# Unzip the uploaded file
for fn in uploaded.keys():
  if fn.endswith('.zip'):
    with zipfile.ZipFile(fn, 'r') as zip_ref:
      zip_ref.extractall('.')
    print(f"Extracted {fn}")

# Assuming the CSV is now extracted as 'smart_city_traffic_stress_dataset.csv'
df = pd.read_csv("smart_city_traffic_stress_dataset.csv")

print("Dataset Loaded Successfully")

print(df.shape)

df.head()
print("="*60)
print("Dataset Information")
print("="*60)

df.info()
df.describe(include='all').T
missing = pd.DataFrame(df.isnull().sum(),columns=['Missing Values'])

missing['Percentage'] = (missing['Missing Values']/len(df))*100

missing.sort_values(by='Percentage',ascending=False)
duplicates = df.duplicated().sum()

print("Duplicate Rows :",duplicates)

df = df.drop_duplicates()

print("Shape After Removing Duplicates")

print(df.shape)

# The original dataset had 'timestamp' for feature engineering, but the new dataset does not.
# This section is removed as the columns 'timestamp', 'day_of_week', 'hour', 'truck_count', 'bus_count', 'car_count', 'bike_count', 'occupancy_rate', and 'congestion_level' are not present.
# The focus now seems to be predicting 'stress_index'.

print(df.columns)

print(df.shape)

target='stress_index'
# drop_columns=['timestamp'] # 'timestamp' is not in the dataset anymore
# No columns to drop that are not the target or features based on the new dataset structure.

x=df.drop(columns=target)
y=df[target]
print(x.shape)
print(y.shape)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print(x_train.columns)
print(x_test.columns)
# print(y_train.columns) # y is a Series, doesn't have columns attribute
# print(y_test.columns) # y is a Series, doesn't have columns attribute
categorical=df.select_dtypes(include='object').columns

print(categorical)
encoder_dict={}

for col in categorical:

    le=LabelEncoder()

    df[col]=le.fit_transform(df[col])

    encoder_dict[col]=le

print("Encoding Completed")
df.head()
X=df.drop(target,axis=1)

y=df[target]

print(X.shape)

print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(

    X,
    y,
    test_size=0.20,
    random_state=42

)
print("Training Samples :",X_train.shape)

print("Testing Samples :",X_test.shape)
print("Training Samples :",y_train.shape)

print("Testing Samples :",y_test.shape)
print("="*60)
print("Training Baseline LightGBM Model")
print("="*60)

baseline_model = LGBMRegressor(
    random_state=42,
    objective='regression'
)

baseline_model.fit(X_train, y_train)

baseline_pred = baseline_model.predict(X_test)

print("Baseline Model Training Completed")

# ==========================================================
# BASELINE MODEL EVALUATION
# ==========================================================

baseline_r2 = r2_score(y_test, baseline_pred)
baseline_mae = mean_absolute_error(y_test, baseline_pred)
baseline_rmse = np.sqrt(mean_squared_error(y_test, baseline_pred))
baseline_mape = mean_absolute_percentage_error(y_test, baseline_pred)

print("="*60)
print("BASELINE MODEL PERFORMANCE")
print("="*60)

print(f"R² Score : {baseline_r2:.4f}")
print(f"MAE      : {baseline_mae:.4f}")
print(f"RMSE     : {baseline_rmse:.4f}")
print(f"MAPE     : {baseline_mape:.4f}")

# ==========================================================
# OPTUNA HYPERPARAMETER TUNING
# ==========================================================

def objective(trial):
    params = {
        'objective': 'regression',
        'metric': 'rmse',
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000, step=100),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.1),
        'num_leaves': trial.suggest_int('num_leaves', 20, 200),
        'max_depth': trial.suggest_int('max_depth', 5, 20),
        'subsample': trial.suggest_uniform('subsample', 0.7, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.7, 1.0),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 50),
        'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-3, 10.0),
        'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-3, 10.0),
        'random_state': 42,
        'n_jobs': -1
    }

    model = LGBMRegressor(**params)
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, prediction))
    return rmse

print("Starting Optuna Hyperparameter Tuning...")
print("Running Optuna trials...") # Added print statement
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50, show_progress_bar=False) # Removed show_progress_bar

print("Hyperparameter Tuning Completed with Optuna")
print("="*60)
print("BEST PARAMETERS (Optuna)")
print("="*60)

print(study.best_params)

# ==========================================================
# FINAL MODEL (Optuna Tuned)
# ==========================================================

best_model = LGBMRegressor(**study.best_params, random_state=42, objective='regression')
best_model.fit(X_train,y_train)

prediction = best_model.predict(X_test)

print("Final Model Trained Successfully with Optuna Tuned Parameters")

# ==========================================================
# FINAL MODEL PERFORMANCE (Optuna Tuned)
# ==========================================================

r2 = r2_score(y_test,prediction)
mae = mean_absolute_error(y_test,prediction)
rmse = np.sqrt(mean_squared_error(y_test,prediction))
mape = mean_absolute_percentage_error(y_test,prediction)

print("="*60)
print("FINAL MODEL PERFORMANCE (Optuna Tuned)")
print("="*60)

print(f"R² Score  : {r2:.5f}")
print(f"MAE       : {mae:.5f}")
print(f"RMSE      : {rmse:.5f}")
print(f"MAPE      : {mape:.5f}")
from sklearn.model_selection import cross_val_score

scores = cross_val_score(

    best_model,

    X,

    y,

    cv=5,

    scoring='r2',

    n_jobs=-1

)

print("="*60)
print("CROSS VALIDATION")
print("="*60)

print(scores)

print("Average R² :",scores.mean())
# ==========================================================
# FEATURE IMPORTANCE
# ==========================================================

importance = pd.DataFrame({

    'Feature':X.columns,

    'Importance':best_model.feature_importances_

})

importance = importance.sort_values(

    by='Importance',

    ascending=False

)

plt.figure(figsize=(12,8))

sns.barplot(

    x='Importance',

    y='Feature',

    data=importance

)

plt.title("LightGBM Feature Importance")

plt.show()
# ==========================================================
# SAVE MODEL
# ==========================================================

pickle.dump(

    best_model,

    open("traffic_stress_lightgbm.pkl","wb")

)

print("Model Saved Successfully")
pickle.dump(

    encoder_dict,

    open("label_encoders.pkl","wb")

)

print("Encoders Saved Successfully")
sample_prediction = best_model.predict(X_test.iloc[:5])

print("Sample Predictions")

print(sample_prediction)
prediction = best_model.predict(X_test)
plt.figure(figsize=(10,6))

plt.scatter(
    y_test,
    prediction,
    alpha=0.6
)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    color="red",
    linewidth=2
)

plt.xlabel("Actual Stress Index")
plt.ylabel("Predicted Stress Index")
plt.title("Actual vs Predicted Stress Index")

plt.grid(True)

plt.show()
residual = y_test - prediction

plt.figure(figsize=(10,6))

plt.scatter(
    prediction,
    residual,
    alpha=0.5
)

plt.axhline(
    0,
    color='red',
    linestyle='--'
)

plt.xlabel("Predicted Values")

plt.ylabel("Residual")

plt.title("Residual Plot")

plt.show()
plt.figure(figsize=(10,6))

sns.histplot(
    residual,
    bins=40,
    kde=True
)

plt.title("Residual Distribution")

plt.show()
importance = pd.DataFrame({

    "Feature":X.columns,

    "Importance":best_model.feature_importances_

})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

plt.figure(figsize=(12,8))

sns.barplot(

    data=importance.head(20),

    x="Importance",

    y="Feature"

)

plt.title("Top 20 Important Features")

plt.show()
error = abs(y_test - prediction)

plt.figure(figsize=(10,6))

plt.plot(error.values)

plt.title("Prediction Error")

plt.xlabel("Samples")

plt.ylabel("Absolute Error")

plt.show()
plt.figure(figsize=(10,6))
sns.histplot(error,bins=40,kde=True)
plt.title("prediction error distibution")
plt.show()
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(

    best_model,

    X,

    y,

    cv=5,

    scoring="r2",

    train_sizes=np.linspace(0.1,1.0,5),

    n_jobs=-1

)

train_mean = train_scores.mean(axis=1)

test_mean = test_scores.mean(axis=1)

plt.figure(figsize=(10,6))

plt.plot(train_sizes,train_mean,label="Training")

plt.plot(train_sizes,test_mean,label="Validation")

plt.legend()

plt.xlabel("Training Samples")

plt.ylabel("R² Score")

plt.title("Learning Curve")

plt.grid(True)

plt.show()

# The dataset name changed to "processed_traffic_stress_dataset.csv" for consistency
df.to_csv(
    "processed_traffic_stress_dataset.csv",
    index=False
)

print("Processed Dataset Saved")
print("all done")
import joblib

joblib.dump(

    best_model,

    "traffic_stress_prediction_model.pkl"

)

print("Model Saved")
model = joblib.load(
    "traffic_stress_prediction_model.pkl"
)
print("Model loaded")
# Decode weather_condition from X_test back to original labels
weather_le = encoder_dict['weather_condition']
x_test_copy = X_test.copy()
x_test_copy['weather_condition_decoded'] = weather_le.inverse_transform(x_test_copy['weather_condition'])

# Create a DataFrame for evaluation
eval_df = pd.DataFrame({
    'actual': y_test,
    'predicted': prediction,
    'weather_condition': x_test_copy['weather_condition_decoded']
})

# Group by weather_condition and calculate metrics
metrics_by_weather = eval_df.groupby('weather_condition').apply(
    lambda x: pd.Series({
        'R2_Score': r2_score(x['actual'], x['predicted']),
        'MAE': mean_absolute_error(x['actual'], x['predicted'])
    })
)

print("="*60)
print("MODEL PERFORMANCE BY WEATHER CONDITION")
print("="*60)
display(metrics_by_weather)
```
